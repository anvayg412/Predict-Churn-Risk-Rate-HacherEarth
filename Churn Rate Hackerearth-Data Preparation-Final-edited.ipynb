```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train=pd.read_csv(r"C:\Users\anvay\Downloads\98efc33085a711eb\dataset\train.csv")
test=pd.read_csv(r"C:\Users\anvay\Downloads\98efc33085a711eb\dataset\test.csv")
```

```python

for i in ['age', 'days_since_last_login', 'avg_time_spent','avg_transaction_value', 'points_in_wallet', 'churn_risk_score']:
    q3,q1=np.percentile(train[i],[75,25])
    iqr=q3-q1
    iqrmin=q1-(1.5*iqr)
    iqrmax=q3+(1.5*iqr)
    countsmin=train.loc[(train[i]<=iqrmin)].index
    countsmax=train.loc[(train[i]>=iqrmax)].index
    train.drop(countsmin,inplace=True,axis=0)
    train.drop(countsmax,inplace=True,axis=0)
```

```python
train.loc[(train['avg_time_spent']>=0)&(train['avg_time_spent']<500)&(train['points_in_wallet'].isnull()==True),['points_in_wallet']]=300.05139
train.loc[(train['avg_time_spent']>=500)&(train['avg_time_spent']<1000)&(train['points_in_wallet'].isnull()==True),['points_in_wallet']]= 701.87
train.loc[(train['avg_time_spent']>=1000)&(train['avg_time_spent']<1500)&(train['points_in_wallet'].isnull()==True),['points_in_wallet']]=1135.268196
```

```python
filters1=train.loc[train['points_in_wallet'].isnull()==True].index
train.drop(filters1,axis=0,inplace=True)
```

```python
def fillNan(df, col, value):
    df[col].fillna(value, inplace=True)
```

```python
fillNan(train, 'preferred_offer_types', train['preferred_offer_types'].mode()[0])
fillNan(test, 'preferred_offer_types', train['preferred_offer_types'].mode()[0])
```

```python
fillNan(train, 'region_category', train['region_category'].mode()[0])
fillNan(test, 'region_category', test['region_category'].mode()[0])
```
